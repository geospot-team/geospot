```python
class Position:
    def __init__(self, lat, lon):
        self.lat = math.pi * lat / 180
        self.lon = math.pi * lon / 180
def get_area(venue, side=0.1, radius=6378,
            corner=Position(59.829833, 30.178096)):
    dy = side / (2 * math.pi * radius)
    i = int(math.floor((venue.lat - corner.lat) / dy))
    lat = corner.lat + dy * i
    dx = side / radius * math.cos(lat)
    j = int(math.floor((venue.lon - corner.lon) / dx))
    return i, j
def get_center(i,j,side = 0.1,radius=6378,
            corner=Position(59.829833, 30.178096)):
    dy = side / (2 * math.pi * radius)
    lat = corner.lat + dy*(i + 0.5) 
    dx = side / radius * math.cos(corner.lat + dy * i)
    lon = corner.lon + dx * (j+0.5)
    return lat * 180 / math.pi,lon * 180 / math.pi

def sphere_dist(x,y,radius = 6378):   
    x_lat = x[0]*math.pi/180
    x_lon = math.pi*x[1]/180
    y_lat = math.pi*y[0]/180
    y_lon = math.pi*y[1]/180
    d_lon = abs(x_lon - y_lon)
    temp = math.sin(x_lat)*math.sin(y_lat)+math.cos(x_lat)*math.cos(y_lat)*math.cos(d_lon)
    if (temp>1):
        temp = 1
    if (temp < -1):
        temp = -1
    return radius*math.acos(temp)
     
```

### NDCG 

https://www.kaggle.com/wiki/NormalizedDiscountedCumulativeGain

http://en.wikipedia.org/wiki/Discounted_cumulative_gain

```python
gain = lambda rel: 2**rel - 1
discount = lambda i: math.log(i+2)
#ratings — hash or series: raiting[area_id] = rank of this area id 
#pred ranks — rank list of areas id: 0 — best ranking area, 1 — second, …
def dcg(ratings, pred_ranks, k = 10,gain = gain, discount = discount):
    rels = (ratings.shape[0] - ratings[pred_ranks][0:k]).div(ratings.shape[0])
    gains = rels.apply(gain)
    discounts = np.log2(np.arange(k) + 2)
    return gains.div(discounts).sum()

#ratings – should be sorted
def ndcg(ratings,pred_ranks, k = 10, gain = gain,discount = discount):
    best = dcg(ratings,ratings.index,k,gain,discount)
    pred = dcg(ratings,pred_ranks,k,gain,discount)
    return pred/best    
```

### CV function

```python
def cross_validation(index, k = 10, shuffle = True, seed = 1):
    rand = np.random.RandomState(seed)
    ind = rand.choice(index, size = len(index),replace = False) if shuffle else index
    for i in range(0,k):
        training = [x for j,x in enumerate(ind) if j % k != i]
        validation = [x for j,x in enumerate(ind) if j % k == i]
        yield training, validation
```

### CV models with target features from dataset

dataset — dataframe with all features

features — features array to use in models

response — name of response column

models — pandas Series of sklearn models (or any other model with fit() and predict() methods). Series index will be used as column name in result dataframe

```python
def score_model(learn,test,response, model):
    test_ratings = get_ratings(test,response)
    learn_ratings = get_ratings(learn,response)
    model = model.fit(learn.drop(response,axis=1),learn[response]) 
    test_pred = pd.Series(model.predict(test.drop(response,axis=1)),index = test.index).order()     
    learn_pred  = pd.Series(model.predict(learn.drop(response,axis=1)),index = learn.index).order() 
    #model_learn_score = model.score(learn,learn_ratings)
    #model_test_score = model.score(test,test_ratings)
    return 1 - ndcg(learn_ratings,learn_pred.index), 1 - ndcg(test_ratings,test_pred.index) #learn then test ncdg-scores
    
def models_cv_with_features(dataset, features, response, models, k = 10):
    df = dataset[features + [response]]
    result = pd.DataFrame()
    for train_ind, val_ind in cross_validation(df.index,k = k):
        #normalization — want mean 0 and standart deviation 1
        scaler = preprocessing.StandardScaler().fit(df.ix[train_ind])
        learn = pd.DataFrame(scaler.transform(df.ix[train_ind]), columns = df.columns)
        validate = pd.DataFrame(scaler.transform(df.ix[val_ind]), columns = df.columns)
        #learn = df.ix[train_ind]
        #validate = df.ix[val_ind]
        scores = models.apply(lambda model:
                              pd.Series(score_model(learn,validate,response,model),index = ["learn","test"])).stack()
        #models.apply(lambda model: pd.Series(score_model(learn,test,"rating",model),index=["learn","test"])).stack()
        #print(scores)
        result = result.append(scores,ignore_index = True)
    result = result.T
    result.index =  pd.MultiIndex.from_tuples(result.index)
    return result

        
    
```

### Wilcoxon test to df of CV-scores

The Wilcoxon signed-rank test tests the null hypothesis that two related paired samples come from the same distribution. In particular, it tests whether the distribution of the differences x - y is symmetric about zero. It is a non-parametric version of the paired T-test.

```python
from scipy.stats import wilcoxon
```

```python
def wilcoxon_df(df):
    results = pd.DataFrame()
    for first in df.columns.values:
        for second in df.columns.values:
            if (first == second) or (not np.any((df[first] != df[second]))):
                results[(first,first)] = (1,1)
                continue
            results[(first,second)] =  wilcoxon(df[first],df[second])
    #results = results.T
    results.columns = pd.MultiIndex.from_tuples(results.columns)
    results = results.T
    return results.unstack()
            
    
```

### Forward features selection

Действуем жадно. На каждом шаге добавляем ту фичу, которая дает наибольшее улучшение. Для нескольких моделей — выбираем фичи, которые в среднем для всех моделей хоорошо (в качестве оценки — mean между всеми моделями). Для features selection для одной модели — достаточно в models оставить только одну модель

```python
def forward_features_selection(dataset,response,models, k = cv_folds):
    result = pd.DataFrame()
    current = set()
    candidates = set(dataset.columns.drop(response))
    i = 0
    while len(candidates) > 0:
        scores = find_feature(dataset,current,candidates,response,models,k = k)
        #print(scores)
        new_feature = scores.xs("test",level = 1).mean().argmin()
        result[(i,new_feature)] = scores[new_feature]
        print("Best feature on {}'s iteration is {}".format(i,new_feature))
        i += 1
        current = current.union({new_feature})
        candidates.remove(new_feature)
    return result
    
    
def find_feature(dataset,current_features, rest_features, response,models, k = cv_folds):
    result = pd.DataFrame()
    for candidate_feature in rest_features:
        features_subset = list(current_features.union({candidate_feature}))
        scores = models_cv_with_features(dataset,features_subset,response,models,k).stack()#.mean()
        #print(scores)
        result[candidate_feature] = scores
    return result
        
```

### Lasso

Лассо — это линейная модель с $l_1$-регуляризацией. Фактически она обнуляет коэффициенты при некоторых фичах и зависит это обнуление от параметра регуляризации $\Rightarrow$ это еще один вид feature selection. 

Модель: $rating = <\alpha, x>$, где $<,>$ — скалярное произведение (обычная линейная модель)

Оптимизируется: $\sum (y_i - <\alpha,x_i>)^2 + \lambda ||\alpha||_1$

Cетку для $\lambda$ надо выбирать в зависимости от датасета. Параметры по-умолчанию, по которым щас рисуется график — окресность, в которой давалась наименьшая ошибка при CV на том датасете, который использовался для отладки всего этого (пока — pizza)

```python
from sklearn.linear_model import lasso_path
```

```python
response = "rating"
    
def score_lasso(learn, test, coef, response = "rating"):
    test_ratings = get_ratings(test,response)
    learn_ratings = get_ratings(learn,response)  
    test_pred = pd.Series(test.drop(response,axis=1).dot(coef),index = test.index).order()     
    learn_pred  = pd.Series(learn.drop(response,axis=1).dot(coef),index = learn.index).order() 
    return 1 - ndcg(learn_ratings,learn_pred.index), 1 - ndcg(test_ratings,test_pred.index) #learn then test ncdg-scores

    
def lasso_cv(df, response,  alphas, k = 10, max_iter = 1e5):   
    result = pd.DataFrame()
    for train_ind, val_ind in cross_validation(df.index,k = k):
            #normalization — want mean 0 and standart deviation 1
            scaler = preprocessing.StandardScaler().fit(df.ix[train_ind])
            learn = pd.DataFrame(scaler.transform(df.ix[train_ind]), columns = df.columns)
            validate = pd.DataFrame(scaler.transform(df.ix[val_ind]), columns = df.columns)
            alphas_lasso, coefs_lasso, _  = lasso_path(learn.drop(response,axis=1), np.array(learn[response]),
                                                       max_iter=max_iter,return_models=False,fit_intercept = False, 
                                                       alphas = alphas)

            models = pd.DataFrame(coefs_lasso.T, index = alphas_lasso,columns = learn.drop(response,axis=1).columns)        
            scores = models.apply(lambda model: 
                                  pd.Series(score_lasso(learn,validate,model,response),index = ["learn","test"]),axis=1).stack()
            #models.apply(lambda model: pd.Series(score_model(learn,test,"rating",model),index=["learn","test"])).stack()
            #print(scores)
            result = result.append(scores, ignore_index = True)
    return result
```

### Backward elimination

Давайте научимся ходить еще и назад. Теперь будем стартовать с полной модели и выкидывать фичи, после которых получается наилучашая модель
(для линейных моделей говорят правильно выкидывать по $Z-score$ — http://en.wikipedia.org/wiki/Standard_score, но там в качестве loss-function — mean-square-error, а у нас под mean-square только тренируется, а выбор фичей хотим по NCDG)

```python
def backward_elimination(dataset,features,response,models, k = cv_folds):
    result = pd.DataFrame()
    current =  set(dataset.columns.drop(response))
    i = 0
    while len(current) > 1:
        scores = drop_features_scores(dataset,current,response,models,k = k)
        #print(scores)
        removed_feature = scores.xs("test",level = 1).mean().argmin()
        result[(i,removed_feature)] = scores[removed_feature]
        print("Removed feature on {}'s iteration is {}".format(i,removed_feature))
        i += 1
        current = current - {removed_feature}
    return result
    
    
def drop_features_scores(dataset,current_features, response,models, k = cv_folds):
    result = pd.DataFrame()
    for candidate_feature in current_features:
        features_subset = list(current_features - {candidate_feature})
        scores = models_cv_with_features(dataset,features_subset,response,models,k).stack()
        #print(scores)
        result[candidate_feature] = scores
    return result
        
```

### Best subset

score with all combinations and it's $\binom {n} {k}$, where n — amount of features. Let's wait for several hundred years =)


А вообще есть информация о такой штуке (source: Elements of Statical Learning), но я про это ничего особо пока не знаю и может потом посмотрю (да и допустимые 30-40 для нас могут быть 10-15, у нас все-таки нету кластера, на котором все можно было бы посчитать):

Best subset regression finds for each $k ∈ {0, 1, 2, . . . , p}$ the subset of size k that gives smallest residual sum of squares (3.2). An efficient algorithm — the leaps and bounds procedure (Furnival and Wilson, 1974)—makes this feasible for p as large as 30 or 40.

```python
from itertools import combinations
import multiprocessing
from multiprocessing import Pool
```

```python
class Subset:
    def __init__(self, dataset, models, subset, response, k=10):
        self.dataset = dataset
        self.models = models
        self.features = list(subset)
        self.subset = subset
        self.response = response
        self.k = k
```

```python
def score_combination(dataset, features, response, models, comb=2, k=10,threads_count = multiprocessing.cpu_count()):
    # result = pd.DataFrame()
    subsets = [Subset(dataset, models, subset, response, k=k) for subset in combinations(features, comb)]
    subsets_index = [subset.subset for subset in subsets]
    #print(len(subsets_index))
    pool = Pool(threads_count)
    result = pool.map(models_cv_with_features, subsets)
    pool.close()
    pool.join()
    # for subset in combinations(features, comb):
    #     features_subset = list(subset)
    #     scores = models_cv_with_features(dataset, features_subset, response, models, k).mean()
    #     result[subset] = scores
    result = pd.DataFrame(result)
    result.index = subsets_index
    return result.T

```

```python

def models_cv_with_features(subset):
    df = subset.dataset[subset.features + [subset.response]]
    result = pd.DataFrame()
    for train_ind, val_ind in cross_validation(df.index, k=subset.k):
        #normalization — want mean 0 and standart deviation 1
        scaler = preprocessing.StandardScaler().fit(df.ix[train_ind])
        learn = pd.DataFrame(scaler.transform(df.ix[train_ind]), columns=df.columns)
        validate = pd.DataFrame(scaler.transform(df.ix[val_ind]), columns=df.columns)
        #learn = df.ix[train_ind]
        #validate = df.ix[val_ind]
        scores = subset.models.apply(lambda model:
                                     pd.Series(score_model(learn, validate, subset.response, model),
                                               index=["learn", "test"])).stack()
        #models.apply(lambda model: pd.Series(score_model(learn,test,"rating",model),index=["learn","test"])).stack()
        #print(scores)
        result = result.append(scores, ignore_index=True)
    return result.mean()
```

```python
def best_subset(dataset, response, models, max_comb=6, k=cv_folds):
    result = pd.DataFrame()
    for i in range(1, max_comb + 1):
        scores = score_combination(dataset, dataset.columns.drop(response), response, models, comb=i)
        scores.index = pd.MultiIndex.from_tuples(scores.index)
        #print(scores)
        new_feature = scores.xs("test", level=1).mean().argmin()
        result[i] = scores[new_feature]
        print("Best feature on {}'s iteration is {}".format(i, new_feature))
    return result.T
```
